```python
import pandas as pd
import numpy as np

# 1. Loading the raw datasets
steam_df = pd.read_csv(r"C:\Users\dkhan\OneDrive\Desktop\Gaming_Fatigue_Project\02_PREPARE_PHASE\Raw_data\steam_store_games.csv")
sales_df = pd.read_csv(r"C:\Users\dkhan\OneDrive\Desktop\Gaming_Fatigue_Project\02_PREPARE_PHASE\Raw_data\Video_Game_Sales_and_Ratings.csv")
```

```python
# 2. Standerizing Coloumn names

steam_df.columns = steam_df.columns.str.lower().str.replace(' ', '_')
sales_df.columns = sales_df.columns.str.lower().str.replace(' ', '_')
```

```python
# 3. Checking for duplicates

print("--- DUPLICATE CHECK ---")
print(f"Duplicate rows in Steam data : {steam_df.duplicated().sum()}")
print(f"Duplicate rows in Sales data : {sales_df.duplicated().sum()}")
```

```python
# 4. Checking for missing values

print("\n--- MISSING VALUES (SALES) ---")
sales_nulls = sales_df.isnull().sum()
print(sales_nulls[sales_nulls > 0])
```

```python
# 5. Dropping critical missing rows

sales_df = sales_df.dropna(subset =['name', 'year_of_release', 'genre'])
```

```python
# 6. Fixing the 'user_score' text trap

sales_df['user_score'] = sales_df['user_score'].replace('tbd', np.nan)
sales_df['user_score'] = sales_df['user_score'].astype(float)
```

```python
# 7. Cleaning up data types

sales_df['year_of_release'] = sales_df['year_of_release'].astype(int)
```

```python
# 8. Verifying the cleaning

print("--- CLEANING VERIFICATION ---")
print(f"Total rows remaining in Sales data : {len(sales_df)}")
print("\nData Types Check:")
print(sales_df[['year_of_release' , 'user_score' , 'critic_score']].dtypes)
```

```python
# 9. Cleaning Steam Release Dates
# Converting the text string to a proper datetime object

steam_df['release_date'] = pd.to_datetime(steam_df['release_date'])

# extracting just the year into a new coloumn called 'release year'

steam_df['release_year'] = steam_df['release_date'].dt.year
```

```python
# 10. Verifying Steam Cleaning

print("--- STEAM CLEANING VERIFICATION ---")
print(steam_df[['release_date' , 'release_year']].head())
```

```python
#11. Saving the Clean Data

steam_df.to_csv(r"C:\Users\dkhan\OneDrive\Desktop\Gaming_Fatigue_Project\03_PROCESS_PHASE\Converted data\steam_data_clean.csv" , index = False)
sales_df.to_csv(r"C:\Users\dkhan\OneDrive\Desktop\Gaming_Fatigue_Project\03_PROCESS_PHASE\Converted data\sales_data_clean.csv" , index=False)

print("\n-- PROCESS PHASE COMPLETE ---")
print("Clean datasets saved successfully")
```
